```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 12, 10
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression as lr
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import ElasticNet as enet
from sklearn.
```

```python
#Define input array with angles from 60deg to 300deg converted to radians
x = np.array([i*np.pi/180 for i in range(60,300,4)])
np.random.seed(10)  #Setting seed for reproducability
y = np.sin(x) + np.random.normal(0,0.15,len(x))
data = pd.DataFrame(np.column_stack([x]),columns=['x'])
plt.plot(data['x'],y,'.')
```

```python
xt = np.array([i*np.pi/180 for i in range(60,300,1)])
np.random.seed(10)  #Setting seed for reproducability
yt = np.sin(xt) + np.random.normal(0,0.15,len(xt))
```

```python
x=np.array(x).reshape(-1,1)
y=np.array(y).reshape(-1,1)
for i in range(1,16):  #power of 1 is already there
    colname = 'x_%d'%i      #new var will be x_power
    data[colname] = data['x']**i
```

```python
X=data.loc[:,'x_1':'x_15'].values
y=y
```

```python
#Define input array with angles from 60deg to 300deg converted to radians
xt = np.array([i*np.pi/180 for i in range(60,300,1)])
np.random.seed(10)  #Setting seed for reproducability
yt = np.sin(xt) + np.random.normal(0,0.15,len(xt))
datat = pd.DataFrame(np.column_stack([xt]),columns=['xt'])
xt=np.array(xt).reshape(-1,1)
yt=np.array(yt).reshape(-1,1)
for i in range(1,16):  #power of 1 is already there
    colnamet = 'x_%d'%i      #new var will be x_power
    datat[colnamet] = datat['xt']**i
```

```python
Xt=datat.loc[:,'x_1':'x_15'].values
yt=yt
```

```python
model=lr()
model.fit(X,y)
dplot= plt.scatter(x,y,color='blue',label='Sample data')
plt.hold(True)
linplot = plt.plot(xt,model.predict(Xt),color='red',label='Linear regression')



plt.xlabel('x',size=20)
plt.ylabel('$\hat{f}(x)$',size=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)


trueplot = plt.plot(xt,np.sin(xt),color='black',label='True function')
plt.legend(prop={'size': 20})
plt.show()
```
